Q1. What is Flask Framework? What are the advantages of Flask Framework?

Answer :  Flask is a lightweight and flexible web framework for building web applications in Python. It was created by Armin Ronacher and is considered one of the most popular web frameworks in Python. Flask provides a simple and elegant approach to building web applications, making it easy to get started with and highly customizable to meet specific requirements.

Advantages of using Flask Framework:

Lightweight and easy to learn: Flask is a lightweight framework and has a simple and easy-to-learn syntax. It does not have any dependencies on external libraries or tools, making it easy to get started with.

Highly customizable: Flask is highly customizable and can be extended with a variety of third-party libraries to meet more complex requirements.

Built-in development server: Flask comes with a built-in development server, making it easy to test and debug your application.

Good documentation: Flask has excellent documentation, with detailed explanations and examples, making it easy for developers to use and learn from.

URL routing: Flask provides a built-in URL routing system, which allows you to define URLs and associate them with functions that handle requests to those URLs.

Good community support: Flask has a strong community of developers, who actively contribute to the development of the framework, and provide support to other developers who use Flask.

Overall, Flask is a great choice for developers who want to quickly create web applications with Python and is especially useful for projects that require a high degree of customization or have specific requirements that may not be easily met by other web frameworks.

Q3. What is App routing in Flask? Why do we use app routes?

Answer :  App routing in Flask refers to the process of defining URL routes and associating them with specific functions that will handle requests to those routes. In Flask, app routes are defined using the @app.route() decorator, which specifies the URL that the view function should respond to and the HTTP methods that the view function should handle.
